### Apply Oversampling

**⚠ Warning**: In this project, we are using upsampling to address the issue of an unbalanced dataset. However, in real-world scenarios, relying solely on upsampling the minority class can lead to challenges. While this method increases the size of the minority class using existing data from the training dataset, it may introduce predictive bias and increase the risk of overfitting the model to the minority class.

An alternative technique for handling unbalanced datasets is downsampling the majority class while applying upweighting to the downsampled instances. Although it may seem counterintuitive to downscale and upweight the majority class, this approach helps mitigate predictive bias. For more information on this topic, you can refer to [Google's Machine Learning Crash Course on Imbalanced Datasets](https://developers.google.com/machine-learning/crash-course/overfitting/imbalanced-datasets).

## Features in Dataset

### 1. Node Features

Each node in the graph represents an atom within the molecule. The `_get_node_features` function gathers a variety of atomic properties to create a feature vector for each atom:

- **Atomic Number** (`atom.GetAtomicNum()`): Represents the element type of the atom (e.g., Carbon = 6, Oxygen = 8).
- **Degree** (`atom.GetDegree()`): Number of directly bonded neighboring atoms.
- **Formal Charge** (`atom.GetFormalCharge()`): The electric charge assigned to an atom.
- **Hybridization** (`atom.GetHybridization()`): The atomic hybridization state, such as sp, sp2, or sp3, indicating bonding behavior.
- **Aromaticity** (`atom.GetIsAromatic()`): A boolean indicating whether the atom is part of an aromatic ring.
- **Total Number of Hydrogens** (`atom.GetTotalNumHs()`): The number of hydrogen atoms attached to this atom.
- **Radical Electrons** (`atom.GetNumRadicalElectrons()`): The number of unpaired electrons on the atom.
- **In Ring** (`atom.IsInRing()`): A boolean indicating if the atom is part of any ring structure.
- **Chirality** (`atom.GetChiralTag()`): Information on the stereochemistry of the atom, which is relevant for asymmetric centers.

These features collectively describe each atom's chemical properties and help the model learn meaningful patterns related to molecular structure.

### 2. Edge Features

Each edge in the graph represents a bond between two atoms. The `_get_edge_features` function captures information about these bonds:

- **Bond Type (as Double)** (`bond.GetBondTypeAsDouble()`): Encodes bond order, such as single (1.0), double (2.0), triple (3.0), or aromatic (1.5) bonds.
- **In Ring** (`bond.IsInRing()`): A boolean indicating whether the bond is part of a ring structure.

These edge features help to represent the connectivity between atoms, capturing different bond types and whether these bonds are in cyclic structures.

### Adjacency Information

The adjacency information, generated by `_get_adjacency_info`, creates an edge list (`edge_indices`) representing which atoms are bonded to each other. For undirected graphs, each bond is represented in both directions (i.e., from atom `i` to atom `j` and from atom `j` to atom `i`), which is essential for graph-based learning algorithms.


```python
class MoleculeDataset(Dataset):
    def __init__(self, root, filename, test=False, transform=None, pre_transform=None):
        """
        Initializes the MoleculeDataset.

        Args:
            root (str): Directory where dataset is stored, which includes raw_dir and processed_dir.
            filename (str): Name of the raw dataset file.
            test (bool, optional): If True, processes test data. Defaults to False.
            transform (callable, optional): A function/transform to apply to the data. Defaults to None.
            pre_transform (callable, optional): A function/transform to apply to data before it’s saved. Defaults to None.
        """
        self.test = test
        self.filename = filename
        self.invalid_smiles = []
        super().__init__(root, transform, pre_transform)

    @property
    def raw_file_names(self):
        """Returns the raw file names."""
        return [self.filename]

    @property
    def processed_file_names(self):
        """Generates processed file names based on the data."""
        self.data = pd.read_csv(self.raw_paths[0]).reset_index()

        if self.test:
            return [f"data_test_{i}.pt" for i in self.data.index if i not in self.invalid_smiles]
        else:
            return [f"data_{i}.pt" for i in self.data.index if i not in self.invalid_smiles]

    def download(self):
        """Override this method to implement dataset downloading."""
        raise NotImplementedError("Download function is not implemented.")

    def process(self):
        """Processes the raw data and stores it in the processed directory."""
        self.data = pd.read_csv(self.raw_paths[0])

        for index, mol in tqdm(self.data.iterrows(), total=self.data.shape[0]):
            mol_obj = Chem.MolFromSmiles(mol['smiles'])

            # Extract molecular features
            node_features = self._get_node_features(mol_obj)
            edge_features = self._get_edge_features(mol_obj)
            edge_index = self._get_adjacency_info(mol_obj)
            label = self._get_labels(mol["HIV_active"])

            # Create a Data object
            data = Data(
                x=node_features,
                edge_index=edge_index,
                edge_attr=edge_features,
                y=label,
                smiles=mol['smiles']
            )

            # Save the data object to the processed directory
            processed_file = f"data_test_{index}.pt" if self.test else f"data_{index}.pt"
            torch.save(data, os.path.join(self.processed_dir, processed_file))

    def _get_node_features(self, mol):
        """Extracts node features (atomic properties) from the molecule."""
        node_feats = []
        for atom in mol.GetAtoms():
            features = [
                atom.GetAtomicNum(),
                atom.GetDegree(),
                atom.GetFormalCharge(),
                atom.GetHybridization(),
                atom.GetIsAromatic(),
                atom.GetTotalNumHs(),
                atom.GetNumRadicalElectrons(),
                atom.IsInRing(),
                atom.GetChiralTag()
            ]
            node_feats.append(features)

        return torch.tensor(np.array(node_feats), dtype=torch.float)

    def _get_edge_features(self, mol):
        """Extracts edge features (bond properties) from the molecule."""
        edge_feats = []
        for bond in mol.GetBonds():
            features = [
                bond.GetBondTypeAsDouble(),
                bond.IsInRing()
            ]
            # Add edge features for both directions
            edge_feats.extend([features, features])

        return torch.tensor(np.array(edge_feats), dtype=torch.float)

    def _get_adjacency_info(self, mol):
        """Generates adjacency information (edge indices) for the molecule."""
        edge_indices = []
        for bond in mol.GetBonds():
            i, j = bond.GetBeginAtomIdx(), bond.GetEndAtomIdx()
            edge_indices.extend([[i, j], [j, i]])

        return torch.tensor(np.array(edge_indices).T, dtype=torch.long)

    def _get_labels(self, label):
        """Converts the label to a tensor."""
        return torch.tensor([label], dtype=torch.int64)

    def len(self):
        """Returns the length of the dataset."""
        return len(self.data)

    def get(self, idx):
        """Retrieves the data object at the specified index."""
        filename = f"data_test_{idx}.pt" if self.test else f"data_{idx}.pt"
        return torch.load(os.path.join(self.processed_dir, filename), weights_only=False)
```

```python
import deepchem as dc

class MoleculeDatasetV2(Dataset):
    def __init__(self, root, filename, test=False, transform=None, pre_transform=None):
        """
        Args:
            root (str): Directory where the dataset is stored. It contains two subdirectories:
                        - 'raw_dir' for downloading raw data.
                        - 'processed_dir' for storing processed data.
            filename (str): The name of the raw data file (CSV).
            test (bool, optional): Flag to specify if the dataset is for testing. Defaults to False.
            transform (callable, optional): A function/transform to apply on each data sample. Defaults to None.
            pre_transform (callable, optional): A function/transform to apply on the data before saving to disk. Defaults to None.
        """
        self.test = test
        self.filename = filename
        self.invalid_smiles = []  # Track invalid SMILES strings
        super().__init__(root, transform=transform, pre_transform=pre_transform)  # Pass root to parent class

    @property
    def raw_file_names(self):
        """
        List of raw files. If the files exist in raw_dir, download is skipped.
        """
        return [self.filename]

    @property
    def processed_file_names(self):
        """
        List of processed files. If these files exist, processing is skipped.
        Excludes invalid SMILES entries.
        """
        self.data = pd.read_csv(self.raw_paths[0]).reset_index()
        if self.test:
            return [f"data_test_{i}.pt" for i in self.data.index if i not in self.invalid_smiles]
        else:
            return [f"data_{i}.pt" for i in self.data.index if i not in self.invalid_smiles]

    def download(self):
        """
        Download the dataset (not implemented in this version).
        """
        raise NotImplementedError("Download method is not implemented.")

    def process(self):
        """
        Process the raw data into a format suitable for PyTorch Geometric.
        Featurizes the molecules and saves them as PyTorch tensors.
        """
        self.data = pd.read_csv(self.raw_paths[0]).reset_index()
        featurizer = dc.feat.MolGraphConvFeaturizer(use_edges=True)

        for index, row in tqdm(self.data.iterrows(), total=self.data.shape[0]):
            # Featurize molecule
            mol = Chem.MolFromSmiles(row["smiles"])
            if mol is None:  # Handle invalid SMILES
                self.invalid_smiles.append(index)
                continue

            # Use featurizer to convert molecule to PyTorch Geometric graph
            try:
                f = featurizer.featurize(mol)  # Note: featurize, not _featurizer
                if len(f) == 0 or f[0].node_features is None:  # Check if featurization failed
                    self.invalid_smiles.append(index)
                    continue
                data = f[0].to_pyg_graph()  # Convert to PyG Data object
            except Exception as e:
                print(f"Error featurizing SMILES at index {index}: {e}")
                self.invalid_smiles.append(index)
                continue

            # Assign label and SMILES
            data.y = self._get_label(row["HIV_active"])
            data.smiles = row["smiles"]

            # Apply pre-transform if specified
            if self.pre_transform is not None:
                data = self.pre_transform(data)

            # Save the processed data
            file_name = f"data_test_{index}.pt" if self.test else f"data_{index}.pt"
            torch.save(data, os.path.join(self.processed_dir, file_name))

    def _get_label(self, label):
        """
        Converts label into a tensor.
        """
        return torch.tensor([label], dtype=torch.int64)

    def len(self):
        """
        Returns the number of valid samples in the dataset.
        """
        return len(self.data) - len(self.invalid_smiles)

    def get(self, idx):
        """
        Returns a processed sample by index.
        Adjusts for invalid SMILES indices.
        """
        if idx in self.invalid_smiles:
            raise IndexError(f"Index {idx} corresponds to an invalid SMILES string.")
        file_name = f"data_test_{idx}.pt" if self.test else f"data_{idx}.pt"
        return torch.load(os.path.join(self.processed_dir, file_name), weights_only=False)
```

#### Helpful resources and documentation used in building the model:

- **[Difference Between GATConv and TransformerConv – GitHub Discussion](https://github.com/pyg-team/pytorch_geometric/discussions/6129)**

- [BatchNorm1d -- Pytorch Documentation](https://pytorch.org/docs/stable/generated/torch.nn.BatchNorm1d.html)

```python
import torch
import torch.nn.functional as F
from torch import nn
from torch.nn import Linear, BatchNorm1d, ModuleList
from torch_geometric.nn import TransformerConv, TopKPooling
from torch_geometric.nn import global_mean_pool as gap, global_max_pool as gmp


class GNN(nn.Module):
    def __init__(self, feature_size, model_params):
        super().__init__()

        # Unpack model parameters for easier access
        embedding_size = model_params["model_embedding_size"]
        n_heads = model_params["model_attention_heads"]
        self.n_layers = model_params["model_layers"]
        dropout_rate = model_params["model_dropout_ratio"]
        top_k_ratio = model_params["model_top_k_ratio"]
        self.top_k_every_n = model_params["model_top_k_every_n"]
        dense_neurons = model_params["model_dense_neurons"]
        edge_dim = model_params["model_edge_dim"]

        # Initialize layers
        self.conv_layers = ModuleList([])
        self.transf_layers = ModuleList([])
        self.pooling_layers = ModuleList([])
        self.bn_layers = ModuleList([])

        # First TransformerConv and Linear layer
        self.conv1 = TransformerConv(feature_size, embedding_size, heads=n_heads,
                                     dropout=dropout_rate, edge_dim=edge_dim, beta=True)
        self.transf1 = Linear(embedding_size * n_heads, embedding_size)
        self.bn1 = BatchNorm1d(embedding_size)

        # Other layers (Conv, Linear, BatchNorm, Pooling)
        for i in range(self.n_layers):
            self.conv_layers.append(TransformerConv(embedding_size, embedding_size,
                                                    heads=n_heads, dropout=dropout_rate,
                                                    edge_dim=edge_dim, beta=True))
            self.transf_layers.append(Linear(embedding_size * n_heads, embedding_size))
            self.bn_layers.append(BatchNorm1d(embedding_size))

            # Add pooling layers at specified intervals
            if i % self.top_k_every_n == 0:
                self.pooling_layers.append(TopKPooling(embedding_size, ratio=top_k_ratio))

        # Fully connected layers (dense layers)
        self.linear1 = Linear(embedding_size * 2, dense_neurons)
        self.linear2 = Linear(dense_neurons, dense_neurons // 2)
        self.linear3 = Linear(dense_neurons // 2, 1)

    def forward(self, x, edge_attr, edge_index, batch_index):
        # Initial convolution, transformation, and batch normalization
        x = self.conv1(x, edge_index, edge_attr)
        x = torch.relu(self.transf1(x))
        x = self.bn1(x)

        # To hold the intermediate graph representations
        global_representation = []

        for i in range(self.n_layers):
            # Pass through the convolutional layers
            x = self.conv_layers[i](x, edge_index, edge_attr)
            x = torch.relu(self.transf_layers[i](x))
            x = self.bn_layers[i](x)

            # Apply pooling and aggregation at specified intervals
            if i % self.top_k_every_n == 0 or i == self.n_layers - 1:
                x, edge_index, edge_attr, batch_index, _, _ = self.pooling_layers[i // self.top_k_every_n](x, edge_index, edge_attr, batch_index)
                # Append aggregated representations
                global_representation.append(torch.cat([gmp(x, batch_index), gap(x, batch_index)], dim=1))

        # Combine all aggregated representations
        x = sum(global_representation)

        # Fully connected layers with dropout for regularization
        x = torch.relu(self.linear1(x))
        x = F.dropout(x, p=0.8, training=self.training)
        x = torch.relu(self.linear2(x))
        x = F.dropout(x, p=0.8, training=self.training)
        x = self.linear3(x)

        return x
```

```python
model_params = {
    "model_embedding_size": 64,
    "model_attention_heads": 3,
    "model_layers": 4,
    "model_dropout_ratio": 0.4,
    "model_top_k_ratio": 0.5,
    "model_top_k_every_n": 1,
    "model_dense_neurons": 256,
    "model_edge_dim": 2,
}
```

### Train model

```python
def train_one_epoch(epoch, model, train_loader, optimizer, loss_fn):
    # Enumerate over the data
    all_preds = []
    all_labels = []
    running_loss = 0.0
    step = 0
    for _, batch in enumerate(tqdm(train_loader)):
        # Use GPU
        batch.to(device)
        # Reset gradients
        optimizer.zero_grad()
        # Passing the node features and the connection info
        pred = model(batch.x.float(),
                                batch.edge_attr.float(),
                                batch.edge_index,
                                batch.batch)
        # Calculating the loss and gradients
        loss = loss_fn(torch.squeeze(pred), batch.y.float())
        loss.backward()
        optimizer.step()
        # Update tracking
        running_loss += loss.item()
        step += 1
        all_preds.append(np.rint(torch.sigmoid(pred).cpu().detach().numpy()))
        all_labels.append(batch.y.cpu().detach().numpy())
    all_preds = np.concatenate(all_preds).ravel()
    all_labels = np.concatenate(all_labels).ravel()
    calculate_metrics(all_preds, all_labels, epoch, "train")
    return running_loss/step
```

```python
def test(epoch, model, test_loader, loss_fn):
    all_preds = []
    all_preds_raw = []
    all_labels = []
    running_loss = 0.0
    step = 0
    for batch in test_loader:
        batch.to(device)
        pred = model(batch.x.float(),
                        batch.edge_attr.float(),
                        batch.edge_index,
                        batch.batch)
        loss = loss_fn(torch.squeeze(pred), batch.y.float())

         # Update tracking
        running_loss += loss.item()
        step += 1
        all_preds.append(np.rint(torch.sigmoid(pred).cpu().detach().numpy()))
        all_preds_raw.append(torch.sigmoid(pred).cpu().detach().numpy())
        all_labels.append(batch.y.cpu().detach().numpy())

    all_preds = np.concatenate(all_preds).ravel()
    all_labels = np.concatenate(all_labels).ravel()
    # print(all_preds_raw[0][:10])
    print("Truncated predictions: ", all_preds[:10])
    print("Truncated Labels:      ", all_labels[:10])
    calculate_metrics(all_preds, all_labels, epoch, "test")
    log_conf_matrix(all_preds, all_labels, epoch)
    return running_loss/step
```

```python
import random

# Fix random seeds for reproducibility
torch.manual_seed(42)
np.random.seed(42)
random.seed(42)

# Load the dataset
dataset = MoleculeDatasetV2(root=".", filename="HIV_train_upsampled.csv")
dataset = dataset.shuffle()
split = int(0.8 * len(dataset))
train_dataset = dataset[:split]
test_dataset = dataset[split:]

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Initialize model
model = GNN(feature_size=dataset.num_node_features, model_params=model_params).to(device)

# Optimizer and loss
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_fn = nn.BCEWithLogitsLoss()

epochs = 10
for epoch in range(epochs):
    print(f"\n=== Epoch {epoch} ===")
    train_loss = train_one_epoch(epoch, model, train_loader, optimizer, loss_fn)
    print(f"Train Loss: {train_loss: .4f}")
    test_loss = test(epoch, model, test_loader, loss_fn)
    print(f"Test Loss: {test_loss: .4f}")
```
